```python
datas=pd.DataFrame()
date=['3月27日','3月28日','3月29日']
for nsheet, d in zip(range(1,4),date):
    df=pd.read_excel(r'C:\Users\arthu\Downloads\2021烘焙展銷售 (210408) + 郵局地址.xlsx',sheet_name=nsheet,skiprows=1)
    ncol=df.shape[1]-2 #13
    items=[]
    for i in df.columns:
        if i.startswith('RK') or i.startswith('SA'):
            items.append(i)
    for idx,it in zip(range(2,ncol,2),items):
        print('------------now', idx)
        data=pd.read_excel(r'C:\Users\arthu\Downloads\2021烘焙展銷售 (210408) + 郵局地址.xlsx',sheet_name=nsheet,usecols=[1,idx,idx+1],skiprows=2,skipfooter=4)
        data.columns=['發票','品項','姓名']

        float_to_int=list(data['品項'])
        for f in range(len(float_to_int)):
            if str(float_to_int[f]).endswith('.0') or len(str(float_to_int[f]))==1:
                float_to_int[f]=it+'*'+str(float_to_int[f])[0]
        data['品項']=float_to_int

        if data['姓名'].isna().all():
            print('------done',idx)
            continue
        nta=data['姓名'].str.split('\n',expand=True)
        nta.columns=['姓名','電話','地址']
        #print(nta)
        for n in nta.columns:
            data[n]=nta[n]
        data=data.dropna(subset=['姓名']).reset_index(drop=True)
        data['日期']=d
        print(data)
        datas=datas.append(data)
        
        print('------done',idx)
    print('finish sheet',nsheet)
```

```python
datas.to_excel('data.xlsx',index=False)
```
